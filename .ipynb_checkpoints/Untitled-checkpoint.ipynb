```python
from __future__ import annotations
import torch
from chess_royale_ai import Chessformer, ChessRoyaleGamesProcess
from chess_royale_environment import Game
from chess_royale_environment.boards import RandomBoard

board_size = (9,9)
number_of_games = 8
batch_size = 4
n_descent = 10
load_path = "./weights/self_supervised"

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

chessformer = Chessformer(board_shape=board_size)
chessformer.weakly_load_state_dict(torch.load(load_path))
chessformer.to(device)

process = ChessRoyaleGamesProcess(
    number_of_games=number_of_games, 
    n_descent=n_descent, 
    batch_size=batch_size,
    max_samples_per_game=5,
    always_last_sample=True)
process.start()

while True:
    
    if process.has_request():
        print("request")
        model_inputs, moves, selection_filters = process.get_request()
        model_inputs = [model_input.to(device) for model_input in model_inputs]
        selection_filters = selection_filters.to(device)
        policies, values = chessformer.moves_policies_and_values(model_inputs, moves, selection_filters)
        process.put_response((policies, values))
    if process.has_results():
        result = process.get_results()
        print("a")
        break
```
